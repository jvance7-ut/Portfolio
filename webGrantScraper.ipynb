```python
from bs4 import BeautifulSoup #scraping
import requests #Open URL
import pandas as pd #dataframe
```

```python
#making a GET request
r = requests.get('https://www.samhsa.gov/grants/grants-dashboard?f%5B0%5D=by_state%3AUtah')

print(r)
```

```python
#parsing the HTML
soup = BeautifulSoup(r.content, 'html.parser')
#print(soup.prettify())

#we want the table values
#the HTML value for our table is class_= "table"
#if you need help finding what values you want:
#1. go to the website
#2. right click and select 'Inspect' from the menu
#3. hover over the HTML until ONLY THE AREA YOU WANT is blue
#4. use the HTML tags from that line

table_names = soup.find('table', class_='views-table views-view-table cols-11 display-none desktop:display-block')

# Extract data from the tbody
if table_names:
    # Loop through rows
    for row in table_names.find_all('th'):
      #column names use 'th'
        print(row.text)
    #print(len(row.text))

else:
    print("Table not found")


data_names = []
for row in table_names.find_all('th'):
    data_names.append(row.text.strip())


print("   ")
print(data_names)

print(len(data_names))
```

```python
# Find the table cell values with the specific class using tbody
tbody = soup.find('table', class_='views-table views-view-table cols-11 display-none desktop:display-block')

# Extract data from the tbody
if tbody:
    # Loop through rows
    for row in tbody.find_all('tr'):
        # Loop through cells
        for cell in row.find_all('td'):
            print(cell.text)
    #cell values use 'tr' and 'td'
else:
    print("Tbody not found.")
```

```python
data = []
for row in tbody.find_all('tr'):
    row_data = []
    for cell in row.find_all('td'):
        a_tag = cell.find('a')
        if a_tag and a_tag.has_attr('href'):
            href = a_tag['href']
            text = a_tag.get_text(strip=True)
            row_data.append((text, href))
        else:
              row_data.append(cell.get_text(strip=True))
    data.append(row_data)

df2 = pd.DataFrame(data)
```

```python
r = requests.get("https://www.samhsa.gov/grants/block-grants/sabg")
html_content = r.content
soup = BeautifulSoup(html_content, 'html.parser')

#bullet_points = soup.find_all('li')
#for bullet_point in bullet_points:
#    print(bullet_point.text.strip())

def scrape_bullet_points(url):
    try:
        response = requests.get(url)
        response.raise_for_status()
    except requests.exceptions.RequestException as e:
        print(f"Request error: {e}")
        return None

    soup = BeautifulSoup(response.content, 'html.parser')

    results = []
    for element in soup.find_all(['h1', 'h2', 'h3', 'h4', 'h5', 'h6', 'ul']):
      if element.name in ('h1', 'h2', 'h3', 'h4', 'h5', 'h6'):
        header = element.text.strip()
        results.append({'type': 'header', 'text': header})
      elif element.name == 'ul':
        for li in element.find_all('li'):
          bullet_point = li.text.strip()
          results.append({'type': 'bullet', 'text': bullet_point})

    return results
```

```python
import re

def extract_text_between_matches(text, start_pattern, end_pattern):
    """
    Extracts and prints the text between the first occurrence of a start pattern
    and the first occurrence of an end pattern that follows it.

    Args:
        text: The input string to search within.
        start_pattern: The regular expression pattern for the starting delimiter.
        end_pattern: The regular expression pattern for the ending delimiter.
    """
    start_match = re.search(start_pattern, text)
    if start_match:
        start_index = start_match.end()
        end_match = re.search(end_pattern, text[start_index:])
        if end_match:
            end_index = start_index + end_match.start()
            extracted_text = text[start_index:end_index]
            return extracted_text

```

```python

def grant_fit(text):
  num_rows = 0
  num_yes = 0
  lines = text.splitlines()
  non_blank_lines = [line for line in lines if line.strip()]
  rows = []
  #rows = "\n".join(non_blank_lines)

  for row in non_blank_lines:
    rows.append(row)
    unique_list = list(set(rows))
  if len(non_blank_lines)== 0:
    percent_fit = -1
    #return percent_fit
  else:

    for row in unique_list:
      requirement_input = input("Do you meet the qualifications for: \n" + row + "\n")
      if requirement_input.lower() == "yes" or requirement_input.lower() == "y":
        num_yes = num_yes + 1
        num_rows = num_rows + 1
      else:
        num_rows = num_rows + 1
    if num_rows >= 1:
      percent_fit = (num_yes/num_rows) * 100
    else:
      percent_fit = 0
  return percent_fit
```

```python
iteration = 0
percent_match = [1]
for row in local_df['Link']:

  #print(iteration)
  requirements = pd.DataFrame()
  url = row
  data = scrape_bullet_points(url)

  resultstring = ""
  if data:
      for item in data:
          clean = f"Type: {item['type']}, Text: {item['text']}" + "\n"
          resultstring += clean + "\n"
  else:
      print("Scraping failed or no data found.")

  if "SUBG Grantee Requirements" in resultstring:
    text = resultstring
    start_pattern = "Type: header, Text: SUBG Grantee Requirements"
    end_pattern = r"Type: header, Text:"
    new_value = extract_text_between_matches(text, start_pattern, end_pattern)
    new_val = new_value.replace('Type: bullet, Text: ', '')
    print(new_val)
    gnlp = grant_fit(new_val)
    #percent_match[iteration] = gnlp
    percent_match.append(gnlp)
    #percent_match = percent_match.append(gnlp)
    iteration = iteration + 1
    print(percent_match)




  elif "Performance Requirements" in resultstring:
    text = resultstring
    start_pattern = "Type: header, Text: Performance Requirements"
    end_pattern = r"Type: header, Text:"
    new_value = extract_text_between_matches(text, start_pattern, end_pattern)
    new_val = new_value.replace('Type: bullet, Text: ', '')
    print(new_val)
    gnlp = grant_fit(new_val)
    #percent_match[iteration] = gnlp
    percent_match.append(gnlp)
    #percent_match = percent_match.append(gnlp)
    print(percent_match)



  elif "Eligibility" in resultstring:
    text = resultstring
    start_pattern = "Type: header, Text: Eligibility"
    end_pattern = r"Type: header, Text:"
    new_value = extract_text_between_matches(text, start_pattern, end_pattern)
    new_val = new_value.replace('Type: bullet, Text: ', '')
    gnlp = grant_fit(new_val)
    #percent_match[iteration] = gnlp
    percent_match.append(gnlp)
    #percent_match = percent_match.append(gnlp)
    iteration = iteration + 1
    print(percent_match)



  else:
    new_row = {'name': row, 'col2': "More research required"}
    print(new_row)
    #percent_match[iteration] = -1
    percent_match.append(-1)
    #percent_match = percent_match.append(gnlp)
    iteration = iteration + 1
    print(percent_match)

  print("percent match:", percent_match)





```
